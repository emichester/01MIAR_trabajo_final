````python
def _parse_Location(x):
    if not pd.isna(x):
        return [float(x) for x in "POINT (-72.93361 41.35973)".split("(")[1].split(")")[0].split(" ")]

def _parse_loc(x):
    if type(x) == Location:
        return [x.longitude, x.latitude]
    elif type(x) == int:
        return x
    elif pd.isna(x):
        return 0
    else:
        return [float(x) for x in "POINT (-72.93361 41.35973)".split("(")[1].split(")")[0].split(" ")]
    
def _fill_Location(row):
    if type(row['Location']) == list:
        return row['Location']
    else:
        return row['loc']
````

````python
data['Location'] = data['Location'].apply(_parse_Location)
````
